# Legendre Polynomials

This notebook is intended to provide an application of the tools developed in `legendre_polynomials.py`. It should also provide the theoretical foundation the script is based on.

```python
import legendre_polynomials as lp
import numpy as np

# support of the polynomials
n_points = 200 # number of points to evaluate the polynomials
x = np.linspace(-1, 1, n_points)

n = 8 # maximum order of the polynomial

# the method lp_generator produces numpy array object with shape (n + 1, x)
polynomials = lp.lp_generator(n,x)
polynomials
```

Each row of the output matrix contains a polynomial evaluated over the support. The row number is the degree of the polynomial.
For instance, row 3 represents the polynomial of third degree evauated at each point of the support.

```python
# method that returns the plot of all polynomials from first degree up to maximum degree n over the support x
lp.polynomial_plot(n,x)
```
